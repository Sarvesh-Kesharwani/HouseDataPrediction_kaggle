```python
# pigenhole principle: ie, one customer has multiple transactions
display('no of pigeons/transactions:' ,data['CustomerID'].shape[0])
display('no of holes/customers:' ,data['CustomerID'].nunique())
```

```python
# duplicate rows analysis
print(f'Ture if a row is duplicate in dataset else False : \n{data.duplicated()}\n')
print(f'we have {"%0.2f" % (data.duplicated().sum()/data.shape[0]*100)}% of duplicate rows in data.\n')

# column on left is the index column which is there in all dataframes, It is not considered as column
# thus this below table has only one column, which is legal to pass in loc function 


# since loc function takes index nums(0,1....n) we can pass boolean series and pick specified rows. 
# note: iloc is column wise accessor and loc is row wise accessor
print(f'which are:\n')
display(data.loc[data.duplicated()])

# we don't see any 2 rows duplicate of each other in below table, becoz these selected rows are 2nd or 3rd....
# copies of 1st copy.
```

```python
# dropping duplicate rows
data = data.drop_duplicates(ignore_index=True)
print(f'no of duplicate rows after dropping now are: ' ,data.duplicated().sum(), '\n')
# print(f'now we have only {data.shape[0]} no of rows left after droping duplicates.\n')
```

```python
display(data.info())
```

```python
# percentage of null rows in data
display('Percentage of missing rows in each column:-', (data.isnull().sum()/len(data)*100).sort_values(ascending=False))
print(type((data.isnull().sum()/len(data)*100)))
```

```python
# removing rows where customer_id is null becoz those rows are useless to us, customer_id is primary key here for this analysis_task
data.dropna(subset=['CustomerID'], inplace=True)

# percentage of null rows in data
display('Percentage of missing rows in each column:-', (data.isnull().sum()/len(data)*100).sort_values(ascending=False))
print(type((data.isnull().sum()/len(data)*100)))
```

```python
# picking only the columns which are required for our current analysis_task
print(f'Features in our dataset :\n{data.columns}')
data = data[['CustomerID', 'Invoice No', 'Date', 'Price']]
print(f'\nFeatures in our dataset now:\n{data.columns}')
```

# 1) diagnosing a problem

```python
# next diagnosing the wrong dtype problem in columns
display(data.info())
```

# 2) fixing that diagnosed problem

```python
# as we see col 'Date' has wrong dtype = string, which should be date_type
# converting dtype of col 'Date' from object to date_type 
from pandas import to_datetime 
data['Date'] = to_datetime(data['Date'])
display(data.info())
```

# data Cleansing completed

# now formating the columns to desired format before feeding to the model

```python
# now we need to convert 'Date'_column to calculate the recency_column(ie, our new_desired_format of the existing column 'Date')
# to do that checking the range of dates in the 'Date' column
display(data['Date'].describe())

# to calculate the recency_column we need a refrence_date
# calculating the refrence_col
# refrence date to caculate the recency of the customer
import datetime
refrence_date =  data['Date'].describe()['last']
refrence_date = refrence_date + datetime.timedelta(days = 1)
print(f'refrence_date taken is : {refrence_date}\n')
# as we can see our ref_date if one day ahead of last_date from our dataset




# doing calculation over the 3 existing columns to format theme
Formatted_data = data.groupby('CustomerID').agg({'Date' : lambda CustDates : (refrence_date  - CustDates.max()).days,
                                                 'Invoice No' : lambda invoice : invoice.count(),
                                                 'Price' : lambda amount : amount.sum()})
print('Formatted_data is : ')
display(Formatted_data)

# renaming the existing columns to recency_column, frequency column, and monetry column
Formatted_data.rename(columns={'Date' : 'Recency', 'Invoice No' : 'Frequency', 'Price' : 'Monetry'})
Formatted_data = Formatted_data.reset_index()
display(Formatted_data)
```
