```python
import pandas as pd
train_data = pd.read_csv('train.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

#
f,ax=plt.subplots(1,2,figsize=(10,5))
#
train_data['Survived'].value_counts().plot.pie(explode=[0,0.1], autopct='%1.1f%%', ax=ax[0], shadow=True)
ax[0].set_title('Survived')
ax[0].set_ylabel('')
#
display(train_data['Survived'])
sns.countplot(x=train_data['Survived'], ax=ax[1])
ax[1].set_title('Survived')
plt.show()
```

```python
# multinomial_barGraph
#https://seaborn.pydata.org/generated/seaborn.countplot.html
f,ax=plt.subplots(1,2,figsize=(18,8))

train_data[['Sex','Survived']].groupby(['Sex']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs Sex')

display(train_data)
sns.countplot(x=train_data['Survived'], hue=train_data['Sex'], ax=ax[1])
ax[1].set_title('Sex:Survived vs Dead')

plt.show()
```

```python
# bucketing the continious data of a col in normal_dist using barplot
f,ax=plt.subplots(1,2,figsize=(20,10))

train_data[train_data['Survived']==0].Age.plot.hist(ax=ax[0],bins=20,edgecolor='black',color='red')
ax[0].set_title('Survived= 0')
x1=list(range(0,85,5))
ax[0].set_xticks(x1)

train_data[train_data['Survived']==1].Age.plot.hist(ax=ax[1],color='green',bins=20,edgecolor='black')
ax[1].set_title('Survived= 1')
x2=list(range(0,85,5))
ax[1].set_xticks(x2)

plt.show()

# as we can see in graph b/w 30 and 35 on x-axis, that continious data has been bucketed into bins/buckets and not been taked directly 
```
