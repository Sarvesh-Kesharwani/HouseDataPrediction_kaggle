```python
# importing required libs at a time
# "import matplotlib.pyplot as plt" OR "from matplotlib import pyplot as plt",
# i prefer first method, since it is shorter ie, "import matplotlib.pyplot as plt"
import matplotlib.pyplot as plt

#checking version of matplotlib library, u can't check version of a module ex: pyplot specificly
import matplotlib
print(matplotlib.__version__)

import numpy as np

#preparing x, y dataset
x = np.array([1, 2])
y = np.array([2, 4])
```

```python
#ploting the graph internally and then showing it on screen
plt.plot(x, y)
plt.show()
```

```python
# ploting the graph internally w/o line, just points only 
# and then showing it on screen
plt.plot(x, y, 'o')
plt.show()
```

```python
# ploting the graph internally with multiple points, just points only 
# and then showing it on screen

x = np.array([1,2,5,8,2,3])
y = np.array([0,5,3,2,6,7])
plt.plot(x, y)
plt.show()
```

```python
# ploting the graph internally without x-dataPoints, just points only 
# and then showing it on screen

y = np.array([0,5,3,2,6,7])
plt.plot(y)
plt.show()
# picks sequential x-datapoints from x-axis ex: 0,1,2,3,4,5....and so on
```
